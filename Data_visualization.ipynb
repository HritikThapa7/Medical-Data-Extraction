### Predicting probable diseases according to symptom

Edit the value of the to_check variable with the symptoms that you want to check. Here is the example of the symptom 'fatigue'.

```python
to_check = ' fatigue'           # symptoms to check
```

```python
dn_[to_check]
```

```python
dn_[to_check].diseases
```

```python
dn_[to_check]['count']
```

```python
sum_count = dn_[to_check]['count'].sum()
```

```python
dn_[to_check]['count'].idxmax()
dn_[to_check].loc[[dn_[to_check]['count'].idxmax()]]
```

```python
max_count = dn_[to_check]['count'].max()
dn_[to_check].loc[dn_[to_check]['count'] == max_count]
```

```python
dn_[to_check].loc[dn_[to_check]['diseases'] == 'Acne']['count']
```

```python
dn_[to_check].loc[dn_[to_check]['count'] == max_count].diseases     # for maximum counts only
```

```python
# for all diseases
ds_dict = {}
ds_list = []
for ds in dn_[to_check].diseases:
  ds_dict[ds] = (dn_[to_check].loc[dn_[to_check]['diseases'] == ds]['count'] / sum_count) * 100
  percentage = ((dn_[to_check].loc[dn_[to_check]['diseases'] == ds]['count'] / sum_count) * 100).tolist()
  ds_list.extend(percentage)
```

```python
dn_[to_check]['% count'] = ds_list
```

```python
dn_[to_check].sort_values('count', ascending = False)
```

### To check two or more symptoms

```python
check_symptoms = [' fatigue', ' obesity', ' blackheads']        # The list contains the group of symptoms to be checked.
```

```python
first = True
result = dn_[check_symptoms[0]].copy()
for sym in check_symptoms:
  if first:
    first = False
    continue
  result = result.append(dn_[sym])
result
```

```python
aggregation_functions = {'Symptoms': 'first', 'count': 'sum', '% count': 'sum'}
dn_new = result.groupby(result['diseases'], as_index = False).aggregate(aggregation_functions).copy()

dn_new
```

```python
# for all diseases

ds_combined_list = []
for ds in dn_new.diseases:
  percentage = ((dn_new.loc[dn_new['diseases'] == ds]['count'] / sum_count) * 100).tolist()
  ds_combined_list.extend(percentage)
```

```python
dn_new['% count'] = ds_combined_list
dn_new = dn_new.drop('Symptoms', axis = 'columns').copy()
dn_new
```

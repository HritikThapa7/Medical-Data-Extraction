```python
#Finding attributes like Age and Tests
from spacy.matcher import Matcher
import re 

def find_age(history):
    return {(ent.text.strip()) for ent in nlp_3(history).ents if ent.label_ == "DATE" and re.search('years-old|years old|year-old|year old|yo|YO',ent.text)}

def find_test(test):
    doc_test = nlp_3(test)

    pattern = [{'LIKE_NUM': True},  {"TEXT": {"IN": ["mmHg"]}}]
    pattern2 = []
    matcher = Matcher(nlp.vocab)
    matcher.add("TEST_RESULT", [pattern])

    matches = matcher(doc_test)  

    test_results = {}

    for match_id, start, end in matches:
        string_id = nlp.vocab.strings[match_id] 
        span = doc_test[start:end]      
        test_result = span.text

        if "mmHg" in span.text:
            test_results["BP"] = test_result
    toks = []
    for token in doc_test:
        if token.is_alpha == False:
            toks.append([token.text, token.i])

    for i in range(len(toks)):
        if toks[i][0] == '/':
            test_results["HR"] = toks[i-1][0]
    
    return test_results

# def find_medication()
```

```python
def find_symptoms(h):
    symptoms = set()
#     h_ents = nlp(h).ents
    h2_ents = nlp2(h).ents
#     for ent in h_ents:
#         if ent.label_ == 'PROBLEM':
#               label = 'SYMPTOMS'
#         else:
#               continue
#         data["SYMPTOMS"].add(ent.text)
    for ent in h2_ents:
        if str(ent._.negex) == "False" and str(ent.label_) == "DISEASE":
              symptoms.add(ent.text)
    return symptoms
```

```python
#Finding disease
f = open("./disease_list", mode='r')
file = f.readlines()
disease_lst = []

for line in file:
    dis = str(line.strip())
    disease_lst.append(dis.upper())
    
def find_diagnosis(a):
    diagnosis = set()
    diseases = set()
    a_ents = nlp2(a).ents
    for ent in a_ents:
      if str(ent._.negex) == "False" and str(ent.label_) == "DISEASE":
        try:
          if ent.text not in diagnosis:
            if ent.text.upper() in disease_lst:
              diseases.add(ent.text)
            else:
              diagnosis.add(ent.text)
        except:
          if ent.text in disease_lst:
              diseases.add(ent.text)
          else:
              diagnosis.add(ent.text)
  
    b_ents = nlp(a).ents

    for e in b_ents:
        if str(e._.negex) == "False" and str(e.label_) == "PROBLEM":
            if ent.text not in diagnosis:
              if ent.text in disease_lst:
                diseases.add(ent.text)
              else:
                diagnosis.add(ent.text)
    return [diagnosis,diseases]
```
